```python
from numpy import *
from numpy.linalg import norm
import matplotlib.pyplot as plt

X, Y = meshgrid(linspace(-3, 3, 200), linspace(-3, 3, 200))

d1 = [(sin(i), cos(i)) for i in linspace(0, 2 * pi, 20)]
d2 = [(-1, -1), (1, 1), (1, -1), (-1, 1)]

contains = vectorize(
    lambda x: min(norm(x - p, inf)
                  for p in d1) < 1 or min(norm(x - p) for p in d2) < 1,
    signature='(2)->()')

fig, ax = plt.subplots(figsize=(3, 3))
ax.contourf(X, Y, contains(dstack((X, Y))), [0.5, 1], cmap='gray')
ax.set_aspect('equal')
ax.set_xticks(linspace(-3, 3, 7))
ax.set_yticks(linspace(-3, 3, 7))
fig.savefig('Ex6.pdf')
```
